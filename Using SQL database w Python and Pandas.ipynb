# Using a SQL database w/ Python and Pandas

Open a connection to the database. My database is called ‘sqlite_example.db3’. We import the sqlite3 package and use it to connect to the database. Read in some of the data i’m working with liverpool_pre as a csv file.

```python
import sqlite3 
import pandas as pd

import os
os.chdir('\\\\file\\UsersM$\\mwa150\\Home')

con = sqlite3.connect("\\\\file\\UsersM$\\mwa150\\Home\sqlite_example.db3")
df = pd.read_csv('liverpool_pre.csv', encoding='utf-8')
```

Store this data in my SQL database. We can create a SQL table w/ the Pandas method to_sql. This will write the data frame to a table called “LP_PRE_SQL”. The if_exists parameter specifies what to do if a table of the same name already exists.

- ‘replace’ will replace the pre-existing table
- ‘fail’ will return an error (useful if you want to be sure you don’t overwrite a table)
- ‘append’ will add to the table

```python
df.to_sql("LP_PRE_SQL", con, if_exists='replace', index=False) 
```

Run a query and extract the results. This is done w/ the pandas read_sql_query method. For example, if I want to get all the rows where friends was less than 20, I would use:

```python
pd.read_sql_query("SELECT * FROM LP_PRE_SQL WHERE friends < 20 ",con)
```
